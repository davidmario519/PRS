```python
# ë¶ˆí•„ìš”í•œ ê³µë°± ì œê±°

import pandas as pd

df = pd.read_csv('youtube_channels.csv')

# ê³µë°± ì œê±° í•¨ìˆ˜
def clean_empty_strings(column):
    return column.apply(
        lambda x: [item.strip() for item in eval(x) if item.strip()] if isinstance(x, str) else x
    )

# genresì™€ moods ì—´ì—ì„œ ê³µë°± ì œê±°
df['genres'] = clean_empty_strings(df['genres'])
df['moods'] = clean_empty_strings(df['moods'])

# ê²°ê³¼ í™•ì¸
df[['genres', 'moods']].head()

# ì²˜ë¦¬ëœ ë°ì´í„°ë¥¼ ìƒˆë¡œìš´ CSV íŒŒì¼ë¡œ ì €ì¥
df.to_csv('processed_youtube_channels.csv', index=False, encoding='utf-8-sig')

print("ì²˜ë¦¬ëœ ë°ì´í„°ê°€ 'processed_youtube_channels.csv'ë¡œ ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")
```

```python
# python ë¬¸ë²•ì˜ ë¬¸ìì—´ ì œê±°('[]', '''')

df = pd.read_csv('processed_youtube_channels.csv')

# Python ë¬¸ë²•ì„ ì œê±°í•˜ëŠ” í•¨ìˆ˜
def clean_python_syntax(column):
    return column.apply(
        lambda x: ', '.join(eval(x)) if isinstance(x, str) and x.startswith('[') else x
    )

# 'genres'ì™€ 'moods' ì—´ ì²˜ë¦¬
df['genres'] = clean_python_syntax(df['genres'])
df['moods'] = clean_python_syntax(df['moods'])

df.to_csv('py-syntax-cleaned_youtube_channels.csv', index=False, encoding='utf-8-sig')
```

```python
# cpi '-' ë°ì´í„° ì²˜ë¦¬
def cpi_clean(x):
    # ì²˜ë¦¬ ì¡°ê±´
    if pd.isna(x) or x == '-' or x == '':
        return ''  # ê³µë°±ìœ¼ë¡œ ì²˜ë¦¬
    try:
        # ê°’ì´ ìˆ«ìë¼ë©´ ë°˜í™˜
        return float(x)
    except ValueError:
        # ìˆ«ìê°€ ì•„ë‹Œ ê²½ìš° ê³µë°± ë°˜í™˜
        return ''
    
df['cpi'] = df['cpi'].apply(cpi_clean)

df['cpi']
```

## genres, moods ë¦¬ìŠ¤íŠ¸í™”

```python
# ê¸°ì¡´ ë¬¸ìì—´ì´ì—ˆë˜ genresì™€ moods ë°ì´í„°ë¥¼ python ë¦¬ìŠ¤íŠ¸ë¡œ ë³€í™˜

def make_list(x):
    if isinstance(x, str):
        return x.split(', ')
    else:
        return ''
```

## ê³µë°± ì œê±°, ëŒ€ë¬¸ì -> ì†Œë¬¸ì

```python
# í•œê¸€ í…ŒìŠ¤íŠ¸

s = ['ë‚˜ëŠ” í•œê¸€ì´ë‹¤.', 'ê·¸ ë ‡ ë‹¤', 'I\'m (English man~)']

[str.lower(i.replace(' ', '')) for i in s]
```

```python
def clean_data(x):
    if isinstance(x, list):
        return [str.lower(i.replace(' ', '')) for i in x]
    else:
        if isinstance(x, str):
            return str.lower(x.replace(' ', ''))
        else:
            return ''
```

```python
features = ['genres', 'moods']
for feature in features:
    df[feature] = df[feature].apply(clean_data)
```

```python
for i in df.loc[0]['genres']:
    print(i)
```

## Feature soup

```python
def create_soup(x):
    return ' '.join(x['genres']) + ' ' + ' '.join(x['moods'])
df['soup'] = df.apply(create_soup, axis=1)
df['soup']
```

## Vectorize, Normalize ê·¸ë¦¬ê³  Cosine Similarity

### êµ¬ì¡°ë„

- `genres` & `moods`
    - soup ë°ì´í„°ë¥¼ Bag Of Words(BOW)ë¥¼ í™œìš©í•´ ê° í‚¤ì›Œë“œì— ëŒ€í•´ vectorize
    - matrix size: $\text{ì±„ë„ ìˆ˜} \times \text{í‚¤ì›Œë“œì˜ ìˆ˜}$

- `cpi`
    - ì •ê·œí™”ë¥¼ ì§„í–‰í•˜ì—¬ $[0, 1]$ì˜ ê°’ìœ¼ë¡œ ë³€í™˜
    - matrix size: $\text{ì±„ë„ ìˆ˜} \times 1\text{(ì±„ë„ì˜ CPI)}$

ë‘ í–‰ë ¬ì„ `hstack`ì„ í†µí•´ ìˆ˜í‰ë³‘í•©í•˜ì—¬
$$
\text{ì±„ë„ ìˆ˜} \times (\text{í‚¤ì›Œë“œì˜ ìˆ˜} + 1\text{(ì±„ë„ì˜ CPI)})
$$
ì˜ `final marix` ìƒì„± í›„ cosine similarity ê³„ì‚°

```python
# ì¥ë¥´ì™€ ë¶„ìœ„ê¸° Vectorize

from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(df['soup'])
count_matrix  # 497ê°œì˜ ì±„ë„ì— ëŒ€í•´ ìµœëŒ€ 48ê°œì˜ feature(í‚¤ì›Œë“œ)ë“¤ì´ ì¡´ì¬
```

```python
# CPI ì •ê·œí™”

from sklearn.preprocessing import MinMaxScaler

# NaNì´ë‚˜ ê³µë°±ì„ 0ìœ¼ë¡œ ëŒ€ì²´
df['cpi_cleaned'] = df['cpi'].replace('', 0).astype(float)

# ì •ê·œí™”
scaler = MinMaxScaler()
df['cpi_normalized'] = scaler.fit_transform(df[['cpi_cleaned']])
```

```python
# ì¥ë¥´ & ë¶„ìœ„ê¸° matrix + cpi_normalized marix ê²°í•©

from scipy.sparse import hstack
import numpy as np

cpi_matrix = np.array(df['cpi_normalized']).reshape(-1, 1)  # CPI ê°’ì„ 2D ë°°ì—´ë¡œ ë³€í™˜
final_matrix = hstack([count_matrix, cpi_matrix])  # soupì™€ CPIë¥¼ ê²°í•©
```

```python
# ì½”ì‚¬ì¸ ìœ ì‚¬ë„ ê³„ì‚°

from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(final_matrix, final_matrix)
cosine_sim
```

## Playlist ì¶”ì²œ

```python
# ìœ íŠœë¸Œ í”Œë ˆì´ë¦¬ìŠ¤íŠ¸ ì±„ë„ì˜ ì´ë¦„ì„ ì…ë ¥ë°›ìœ¼ë©´ ì½”ì‚¬ì¸ ìœ ì‚¬ë„ë¥¼ í†µí•´ì„œ ê°€ì¥ ìœ ì‚¬ë„ê°€ ë†’ì€ ìƒìœ„ 10ê°œì˜ ì±„ë„ ëª©ë¡ ë°˜í™˜
def get_recommendations(channel_name, cosine_sim=cosine_sim):
    # ì±„ë„ì„ í†µí•´ì„œ ì „ì²´ ë°ì´í„° ê¸°ì¤€ ê·¸ ì±„ë„ì˜ index ê°’ì„ ì–»ê¸°
    idx = indices[channel_name]
    
    # ì½”ì‚¬ì¸ ìœ ì‚¬ë„ ë§¤íŠ¸ë¦­ìŠ¤ (cosine_sim) ì—ì„œ idx ì— í•´ë‹¹í•˜ëŠ” ë°ì´í„°ë¥¼ (idx, ìœ ì‚¬ë„) í˜•íƒœë¡œ ì–»ê¸°
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    # ì½”ì‚¬ì¸ ìœ ì‚¬ë„ ê¸°ì¤€ìœ¼ë¡œ ë‚´ë¦¼ì°¨ìˆœ ì •ë ¬
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # ìê¸° ìì‹ ì„ ì œì™¸í•œ 10ê°œì˜ ì±„ë„ì„ ìŠ¬ë¼ì´ì‹±
    sim_scores = sim_scores[1:11]
    
    print(sim_scores)  # ì½”ì‚¬ì¸ ìœ ì‚¬ë„ ì¶œë ¥ ì¶œë ¥
    
    # ì¶”ì²œ ì±„ë„ ëª©ë¡ 10ê°œì˜ ì¸ë±ìŠ¤ ì •ë³´ ì¶”ì¶œ
    channel_indices = [i[0] for i in sim_scores]
    
    # ì¸ë±ìŠ¤ ì •ë³´ë¥¼ í†µí•´ ì±„ë„ ì´ë¦„ ì¶”ì¶œ
    return df['channel_name'].iloc[channel_indices]
```
